```python
restricoes_map = {
    1: "Lactose",
    2: "Glúten",
    3: "Soja",
    4: "Ovo",
    5: "Amendoim",
    6: "Vegetarianismo",
    7: "Veganismo",
    8: "Não tenho restrição",
    9: "Outro"
}

objetivo_map = {
    1: "Quero dar a minha opinião sobre produtos",
    2: "Quero descobrir novos produtos para comprar"
}

faixa_renda_map = {
    6: "R$0 - 3.000",
    7: "R$3.000 - 8.000",
    8: "R$8.000 - 16.000",
    9: "R$16.000 - 24.000",
    10: "Superior a R$24.000"
}

# Transformação da coluna 'restricoes_alimentares' usando JSON e o mapeamento
base_gostos['restricoes_alimentares'] = base_gostos['restricoes_alimentares'].apply(
    lambda x: ", ".join([restricoes_map[id_] for id_ in json.loads(x)["restricoes"]]) 
    if pd.notnull(x) and x.startswith("{") else x
)

# Transformação das colunas 'objetivo' e 'faixa_renda' usando o .loc
base_gostos.loc[base_gostos['objetivo_id'].notnull(), 'objetivo_id'] = base_gostos['objetivo_id'].map(objetivo_map)
base_gostos.loc[base_gostos['renda_id'].notnull(), 'renda_id'] = base_gostos['renda_id'].map(faixa_renda_map)
```

```python
base_gostos['data_nascimento'] = pd.to_datetime(base_gostos['data_nascimento'], format='%d/%m/%Y', errors='coerce')

# Calcular a idade com base na data atual
today = datetime.now()
base_gostos['idade'] =  base_gostos['data_nascimento'].apply(lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)) if pd.notnull(x) else None)
```

```python
bins = [0, 19, 59, 120]  # Limites de idade para Jovens, Adultos e Idosos
labels = ["Jovens", "Adultos", "Idosos"]  # Rótulos para as faixas etárias

# Criar a coluna 'faixa_etaria' com as faixas etárias
base_gostos['faixa_etaria'] = pd.cut(base_gostos['idade'], bins=bins, labels=labels, right=True)
```

```python


# 2. Dividir a tabela em dimensões e fato
dim_demografico_cliente= base_nova[[
    'id_demografico_base', 'nome', 'sobrenome', 'data_nascimento', 'genero',
    'cidade', 'estado', 'bairro', 'idade', 'faixa_etaria'
]]


dim_gosto_cliente = base_nova[[
    'id_gosto_base', 'moda', 'pessoa_artistica', 'maquiagem', 'brinquedos_ou_jogos', 
    'esportes', 'tem_pet', 'decorar_casa', 'alimento_saudavel', 'gastronomia', 
    'carro', 'bebida_alcoolica', 'cuidados_pessoais', 'tecnologia', 
    'mae_ou_pai', 'influencer', 'restricoes_alimentares',
]]

fato_cliente = base_nova[[
    'id_fato_base', 'id_demografico_base', 'id_gosto_base', 'data_inscricao', 
    'rede_social', 'renda_id', 'objetivo_id', 'tem_cupom',  'assinantes'
]]
fato_cliente.rename(columns={'renda_id':'faixa_renda', 'objetivo_id': 'objetivo'}, inplace=True)

# 4. Conectar ao banco de dados SQLite
db_path = "Experimentai.db"
conn = sqlite3.connect(db_path)

# 5. Criar as tabelas no banco
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_demografico_cliente (
    id_demografico_base INTEGER PRIMARY KEY,
    nome TEXT,
    sobrenome TEXT,
    data_nascimento TEXT,
    genero TEXT,
    cidade TEXT,
    estado TEXT,
    bairro TEXT,
    idade INTEGER,
    faixa_etaria TEXT
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_gosto_cliente (
    id_gosto_base INTEGER PRIMARY KEY,
    moda BOOLEAN,
    pessoa_artistica BOOLEAN,
    maquiagem BOOLEAN,
    brinquedos_ou_jogos BOOLEAN,
    esportes BOOLEAN,
    tem_pet BOOLEAN,
    decorar_casa BOOLEAN,
    alimento_saudavel BOOLEAN,
    gastronomia BOOLEAN,
    carro BOOLEAN,
    bebida_alcoolica BOOLEAN,
    cuidados_pessoais BOOLEAN,
    tecnologia BOOLEAN,
    mae_ou_pai BOOLEAN,
    influencer BOOLEAN,
    restricoes_alimentares BOOLEAN
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS fato_cliente (
    id_fato_base INTEGER PRIMARY KEY,
    id_demografico_base INTEGER,
    id_gosto_base INTEGER,
    data_inscricao TEXT,
    rede_social TEXT,
    faixa_renda VARCHAR(20),
    objetivo VARCHAR(30),
    tem_cupom BOOLEAN,
    assinantes BOOLEAN,
    FOREIGN KEY(id_demografico_base) REFERENCES dim_demografico_cliente(id_demografico_base),
    FOREIGN KEY(id_gosto_base) REFERENCES dim_gosto_cliente(id_gosto_base)
);
''')

# 6. Inserir os dados nas tabelas
dim_demografico_cliente.to_sql('dim_demografico_cliente', conn, if_exists='append', index=False)
dim_gosto_cliente.to_sql('dim_gosto_cliente', conn, if_exists='append', index=False)
fato_cliente.to_sql('fato_cliente', conn, if_exists='append', index=False)

# Fechar a conexão
conn.commit()
conn.close()

print("Modelo estrela para fato_base_nova foi criado e os dados foram inseridos com sucesso!")
```
